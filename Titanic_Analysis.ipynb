# Titanic Dataset Analysis using Python

### Overview
This notebook analyzes the Titanic dataset, cleaning the data, creating visualizations, and building an interactive dashboard using Dash and Plotly.

### Libraries Used:
- `pandas` → Data manipulation
- `plotly` → Interactive visualizations
- `dash` → Web-based dashboard

```python
import pandas as pd

# Load dataset
df = pd.read_excel("Titanic Dataset.xlsx", engine="openpyxl")

# Preview data
df.head()
```

### Data Cleaning
- Fill missing values in Age with median
- Drop Cabin column (too many missing values)
- Convert categorical data to numeric

```python
df['Age'] = df['Age'].fillna(df['Age'].median())
df = df.drop(columns=['Cabin'])
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})
df['Embarked'] = df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})

# Confirm changes
df.info()
```

## Visualizing Titanic Data
Let's create multiple charts using Plotly to better understand survival patterns.

```python
import plotly.express as px

# Age Distribution
fig1 = px.histogram(df, x='Age', title='Age Distribution')
fig1.show()

# Survival Rate by Class
fig2 = px.bar(df, x='Pclass', y='Survived', color='Pclass', title='Survival Rate by Class')
fig2.show()

# Survival Rate by Gender
fig3 = px.pie(df, names='Sex', values='Survived', title='Survival Rate by Gender')
fig3.show()
```

### Key Insights
- **Age Distribution:** Most passengers are between 20–40 years old.
- **Survival Rate:** Higher in **1st class** and **among females**.
- **Fare vs Age Comparison:** Older passengers tend to pay higher fares.
